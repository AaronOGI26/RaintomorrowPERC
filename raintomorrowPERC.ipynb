```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from keras.models import model_from_json
import matplotlib.pyplot as plt

# cargamos las 4 combinaciones de las compuertas XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# y estos son los resultados que se obtienen, en el mismo orden
y = np.array([[0],[1],[1],[0]], "float32")

epocas=int(input('Introduce el nmero de pocas de entrenamiento para esta red: '))
#Diseo mi modelo de red, podis probarlo en el playground.tensorflow
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#Compilamos el modelo seleccionando la funcin de perdida, el optimizador y la mtrica de evaluacin
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])


#Entrenamos el modelo compilado con el nmero de pocas
model.fit(X, y, epochs=epocas,verbose=False)
modelo=model.fit(X, y, epochs=epocas,verbose=False)

# evaluamos el modelo
scores = model.evaluate(X, y)

#Con esto podis imprimir y mostrar las mtricas del modelo
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

print (model.predict(X).round())
plt.plot(modelo.history['loss'])
```

```python
# Hago series el modelo para JSON, genera un archivo JSON con la arquitectura de la RED.
model_json=model.to_json()
with open ("model.json","w") as json_file:
  json_file.write(model_json)
```

```python
# Hago series los pesos(weights) para HDF5, genera un archivo HDF5 con los pesos obtenidos en el entrenamiento.
model.save_weights("model.h5")
print("Modelo guardado en el PC")
```

```python
# Cargo el json y creo el modelo
json_file=open('model.json','r')
loaded_model_json=json_file.read()
json_file.close()
loaded_model=model_from_json(loaded_model_json)
```

```python
#Evaluo el modelo cargado con los datos de los test
loaded_model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
score=loaded_model.evaluate(X,y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))
# se evalua el modelo cargado con los datos de los test
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
score = loaded_model.evaluate(X, y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))
```
